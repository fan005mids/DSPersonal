```python
# Crawl


import requests
from bs4 import BeautifulSoup

url = 'https://ue.berkeley.edu/students/my-major-map'

reqs = requests.get(url)

soup = BeautifulSoup(reqs.text, 'html.parser')

urls = []
for link in soup.find_all('a'):
    urls.append(link.get('href'))
urls
```

```python
import re
urls_filtered_list = []
for u in urls:
    for i in re.findall('https://ue.berkeley.edu/sites/default/files/.+', str(u)):
        if i not in urls_filtered_list:
            print(i)
            urls_filtered_list.append(i)
urls_filtered_list
```
