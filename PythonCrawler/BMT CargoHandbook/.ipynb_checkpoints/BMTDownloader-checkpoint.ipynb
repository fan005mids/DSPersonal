```python
# Crawl


import requests
from bs4 import BeautifulSoup

url = 'https://www.cargohandbook.com/Category:Products'

reqs = requests.get(url)
#encoding = "utf-8"
soup = BeautifulSoup(reqs.content, 'html.parser')

urls = []
for link in soup.find_all('a'):
    urls.append(link.get('href'))
urls
```

```python
import re
urls_filtered_list = []
for u in urls:
    for i in re.findall('^/.+', str(u)):
        if i not in urls_filtered_list:
            urls_filtered_list.append(i)
            
for u in urls_filtered_list[:]:
    if (len(re.findall("^/Special.+", u)) != 0) or (len(re.findall("^/index.+", u)) != 0):
        urls_filtered_list.remove(u)  
urls_filtered_list = urls_filtered_list[1:-3]
final_urls += urls_filtered_list
final_urls
```

```python
from PyPDF2 import PdfMerger, PdfReader
from pyhtml2pdf import converter

def download_file(download_url, filename):
    converter.convert(download_url, filename, print_options={"displayHeaderFooter": True})
```

```python
mergedObject = PdfMerger()
for fileNumber in range(len(final_urls)):
    pdf_path = final_urls[fileNumber]
    file_str = str(fileNumber)+ '.pdf'

    download_file(pdf_path, file_str)
                
    mergedObject.append(PdfReader(file_str, 'rb'))

mergedObject.write("mergedfilesoutput.pdf")
```
