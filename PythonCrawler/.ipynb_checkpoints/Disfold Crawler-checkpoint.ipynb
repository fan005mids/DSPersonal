```python
# Crawl


import requests
from bs4 import BeautifulSoup

url = 'https://zh.disfold.com/%e7%be%8e%e5%9b%bd%e9%a1%b6%e7%ba%a7%e9%93%b6%e8%a1%8c/'

result = None
while result is None:
    try:
        reqs = requests.get(url)
        result = True
    except:
         pass

soup = BeautifulSoup(reqs.text, 'html.parser')

urls = []
for link in soup.find_all('a'):
    urls.append(link.get('href'))
```

```python
import re
urls_filtered_list = []
for u in urls:
    for i in re.findall('https://zh.disfold.com/%.+', str(u)):
        if i not in urls_filtered_list:
            print(i)
            urls_filtered_list.append(i)
urls_filtered_list
```

```python
def add_utf(raw_html):
    cleantext = re.sub('<head>', '<head> <meta http-equiv="Content-type" content="text/html; charset=utf-8" /><meta charset="UTF-8" />', raw_html)
    return cleantext

def cleanhtml(raw_html):
    cleantext = re.sub('<.*?>', '', raw_html)
    return cleantext+"\n"

def delete_img(raw_html):
    cleantext = re.sub('<img alt.+?/>', '', raw_html)
    return cleantext

def delete_link(raw_html):
    cleantext = re.sub('<link href.+?/>', '', raw_html)
    return cleantext
```

```python

import pdfkit
import requests
import os

from PyPDF2 import PdfFileMerger, PdfFileReader

mergedObject = PdfFileMerger()
for fileNumber in range(len(urls_filtered_list)):
    file_str = str(fileNumber)+ '.pdf'

    result = None
    while result is None:
        try:
            # connect
            r = requests.get(urls_filtered_list[fileNumber])
            soup = BeautifulSoup(r.text, 'html.parser')
            print(urls_filtered_list[fileNumber])
            result = True
        except:
             pass
    with open("output1.html", "w", encoding="utf-16") as file:
        file.write(delete_link(str(soup)))

    with open("output1.html", "r", encoding="utf-16") as f:
        try:
            pdfkit.from_file(f,file_str, options={'encoding': "utf-16", 'disable-local-file-access': "", "enable-javascript": "", "load-error-handling": "ignore", "load-media-error-handling": "ignore"})
        except:
            pass
                
    mergedObject.append(PdfFileReader(file_str, 'rb'), import_bookmarks=False)

mergedObject.write("mergedfilesoutput.pdf")
```
