```python
final_urls = []
```

```python
url_prefix = "https://www.globaldata.com"
```

```python
# Crawl


import requests
from bs4 import BeautifulSoup

url = "https://www.globaldata.com/companies/top-companies-by-sectors/"

reqs = requests.get(url)
#encoding = "utf-8"
soup = BeautifulSoup(reqs.content, 'html.parser')

urls = []
for link in soup.find_all('a'):
    urls.append(link.get('href'))
urls
```

```python
import re
sub_urls = []
urls_filtered_list = []
for u in urls:
    for i in re.findall('^\/companies\/top-companies-by-sector\/.+?\/', str(u)):
        if i not in sub_urls:
            sub_urls.append(i)
            
for sub_u in sub_urls:
    urls_filtered_list.append(sub_u)
    reqs = requests.get(url_prefix+sub_u)
    soup = BeautifulSoup(reqs.content, 'html.parser')
    sub_urls = []
    for link in soup.find_all('a'):
        sub_urls.append(link.get('href'))
    for u in sub_urls:
        for i in re.findall('^\/companies\/top-companies-by-sector\/.+', str(u)):
            if i not in urls_filtered_list:
                urls_filtered_list.append(i)
final_urls += urls_filtered_list
final_urls
```

```python
sub_urls = []
urls_filtered_list = []
for u in urls:
    for i in re.findall('^\/companies\/top-companies-by-advisors-investors\/.+?\/', str(u)):
        if i not in sub_urls:
            sub_urls.append(i)
            
for sub_u in sub_urls:
    urls_filtered_list.append(sub_u)
    reqs = requests.get(url_prefix+sub_u)
    soup = BeautifulSoup(reqs.content, 'html.parser')
    sub_urls = []
    for link in soup.find_all('a'):
        sub_urls.append(link.get('href'))
    for u in sub_urls:
        for i in re.findall('^\/companies\/top-companies-by-advisors-investors\/.+', str(u)):
            if i not in urls_filtered_list:
                print(i)
                urls_filtered_list.append(i)
final_urls += urls_filtered_list
final_urls
```

```python
for i in range(len(final_urls)):
    final_urls[i] = url_prefix + final_urls[i]
final_urls
```

```python
from PyPDF2 import PdfMerger, PdfReader
from pyhtml2pdf import converter

# import weasyprint as wp
# from weasyprint import HTML
# import requests

def download_file(download_url, filename):
    converter.convert(download_url, filename, print_options={"paperHeight": 15, "scale": 0.3})
    
#     # Use requests to get the HTML content of the page
#     response = requests.get(download_url)
#     html_content = response.content.decode('utf-8')

#     # Convert the HTML content to a PDF
#     HTML(string=html_content, base_url=download_url).write_pdf(filename)
```

```python
mergedObject = PdfMerger()
for fileNumber in range(len(final_urls)):
    pdf_path = final_urls[fileNumber]
    file_str = str(fileNumber)+ '.pdf'

    download_file(pdf_path, file_str)
                
    mergedObject.append(PdfReader(file_str, 'rb'))

mergedObject.write("mergedfilesoutput.pdf")
```
